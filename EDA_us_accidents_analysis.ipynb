### Start Time

```python
df.columns
```

```python
df.Start_Time
```

```python
df.Start_Time[0]
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time, format="%m/%d/%Y %I:%M:%S %p")
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Some insights from the graphs above:

- most of the accidents occur between 6am to 10am probably because people commute for work mostly around that time
- next highest percentage of accidents occur between 3pm to 7pm

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek ==6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

On sundays the peak number of accidents occur during the afternoons between 10 am and 3pm unlike weekdays

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek ==0]
sns.distplot(mondays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

On Mondays, like other weekdays, the most number of accidents occur between 7am to 10am and between 4pm to 7pm which is most likely the hours people commute for work.

```python
df.Start_Time.dt.year
```

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
sns.distplot(df.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

From the graph plotted for the year 2019, we can infer that the number of accidents occured during the months of April, May, June and July are the lowest compared to the number of accidents in the other months.

In the above, map we have plotted the latitute and logitude points of where the accidents had occureded and we can see that the number of accidents are higer in the cities closer to the coastal lines as the population there is higher than that in the cities in the central region of America.
